```python
# What is the probability of a male customer buying a KP781 treadmill?
matrix = pd.crosstab(index=df['Product'], columns= df['Gender'], margins= True).reset_index()

# P(KP281/Male) = P(KP281 & Male)/P(Male)
P1 = round(matrix.loc[0][2]/matrix.loc[3][2],2)*100

# P(KP281/Female) = P(KP281 & Female)/P(Female)
P2 = round(matrix.loc[0][1]/matrix.loc[3][1],2)*100

# P(KP481/Male) = P(KP481 & Male)/P(Male)
P3 = round(matrix.loc[1][2]/matrix.loc[3][2],2)*100

# P(KP481/Female) = P(KP481 & Female)/P(Female)
P4 = round(matrix.loc[1][1]/matrix.loc[3][1],2)*100

# P(KP781/Male) = P(KP781 & Male)/P(Male)
P5 = round(matrix.loc[2][2]/matrix.loc[3][2],2)*100

# P(KP781/Female) = P(KP781 & Female)/P(Female)
P6 = round(matrix.loc[2][1]/matrix.loc[3][1],2)*100
print(f"{matrix}\n\nP(KP281/Male): {P1}%, P(KP281/Female): {P2}%\nP(KP481/Male): {P3}%, P(KP481/Female): {P4}%\nP(KP781/Male): {P5}%, P(KP781/Female): {P6}%")
```

The probability of a male customer buying KP781 trademill is 3.5 times the probability of female buying the same
